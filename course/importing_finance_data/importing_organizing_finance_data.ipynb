## Importing From Other Financial Web Resources 

### Importing From Alpha Vantage (Need an API Key)

To register to get your own free API KEY head over here : https://www.alphavantage.co/support/#api-key

```python
from alpha_vantage.timeseries import TimeSeries
```

```python
api_key = input("Please input your api key")
ts = TimeSeries(key = api_key,output_format = 'pandas')
```

#### Selecting a Single Stock Information

```python
aapl_av, metadata = ts.get_daily_adjusted('AAPL', outputsize = 'full')
aapl_av
```

```python
msft_av, metadata = ts.get_daily_adjusted('MSFT', outputsize = 'full')
msft_av.head()
```

### Comparing 2 Stocks Prices Closing Price

```python
msft_close = msft_av['4. close']
aapl_close = aapl_av['4. close']

tech_close = pd.concat([aapl_close, msft_close], axis = 1)
tech_close.head()
```

```python
tech_close.columns = ['AAPL', 'MSFT']
```

## Resampling Data

### Resampling to Bussiness Month

the month of april although as of today *(16th of april)* the resample method will show the date of the last business month on that month and show the last non null value of the column

```python
comparison_data.resample('BM').last()
```

### Resampling to Business Year

Similar to the resampling method in the business month the resampling method will show the index what's the last day of a particular date time and will pick the latest business day and for the value of the column it will select the last non null value of each column

```python
comparison_data.resample('BY').last()
```
